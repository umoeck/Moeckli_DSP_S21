##Exercise 3
We now review k-fold cross-validation.
a.)Explain how k-fold cross-validation is implemented.

- In the k-fold cross-validation, the data set is split in k parts of equal size, whereat on part is used as the the validation data set und k-1 parts are used to fit the model. Then the MSE is calcualted on the validation data set. This precedure is repreated k-fold and each time a different part is treated as a validation data set. As a result we get k times the MSE, which is averaged to calculate the the k-fold CV estimate. For example, in a 10fold cross-validation the data set is spilt in 10 parts, 1 part is used as validation data and 9 parts are used to fit the model.


b.) What are the advantages and disadvantages of k-fold cross-
validation relative to:
i. The validation set approach? 
- Advantage: There is tipically less variability in the k-fold validation than in the validation set approach. All the data in k-fold cross-validation is used to fit and validate the the model. Furthermore, the validation set approach can lead to overestimates of the test error rate.
- Disadvantage: The k-fold cross-validation is computationally more expensive


ii. LOOCV?

- Advantage: The k-fold cross-validation is compuatational less expensive than the LOOVC, especially if n is very large (the LOOVC can also bee seen as a special case of k-fold cross-validation). A more important advantage of k-fold CV is that it often gives more accurate estimates of the test error rate than does LOOCV. There is also an advantage concerning the bias-variance trade-off, the k-fold cross validation tends to have less variance than LOOCV

- Disadvantage: The model is trained on a smaller data set (n/k vs. n-1)
